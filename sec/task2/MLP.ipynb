```python
import numpy as np
from sklearn.neural_network import MLPClassifier


X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])

mlp = MLPClassifier(hidden_layer_sizes=(4,), 
                    activation='relu', 
                    solver='adam', 
                    max_iter=2000, 
                    random_state=1)


mlp.fit(X, y)


predictions = mlp.predict(X)

print("Inputs:")
print(X)
print("\nPredicted XOR Outputs:")
print(predictions)
print("\nAccuracy:", mlp.score(X, y) * 100, "%")
```
